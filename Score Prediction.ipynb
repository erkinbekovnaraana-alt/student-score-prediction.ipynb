```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# данные
data = {
    'hours_studied': [1, 2, 3, 4, 5, 6, 7, 8],
    'exam_score': [50, 55, 60, 65, 70, 75, 80, 90]
}

df = pd.DataFrame(data)

# разделяем данные
X = df[['hours_studied']]
y = df['exam_score']

# создаем модель
model = LinearRegression()
model.fit(X, y)

# предсказание
predicted_score = model.predict([[9]])
print("Predicted score for 9 hours:", predicted_score[0])

# график
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Prediction Model')
plt.show()

# БОНУС (насколько влияет учеба)
print("Model coefficient:", model.coef_[0])
```

## Conclusion

The model predicts that students who study more hours will achieve higher exam scores.
This shows a positive relationship between study time and performance.
